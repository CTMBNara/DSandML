```python
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
def f(x, w0, w1):
    return w0 + w1 * x
```

Циклом:

```python
def error(param):
    w0, w1 = param
    sum = 0
    for y, x in zip(data_y, data_x):
        sum += (y - f(x, w0, w1))**2
    return sum
```

Матричными операциями:

```python
def error1(param):
    w0, w1 = param
    v = data_y - f(data_x, w0, w1)
    return np.dot(v, v.T)
```

Исследуем зависимость времени работы каждой из функций от объема данных:

```python
t_array = []
t_array1 = []

for i in range(100, 10000, 100):
    data_x = np.random.rand(i)
    data_y = np.random.rand(i)
    w = np.random.rand(2)
    
    t0 = time.time()
    error(w)
    t = time.time()
    t_array.append(t - t0)
    
    t0 = time.time()
    error1(w)
    t = time.time()
    t_array1.append(t - t0)
```

```python
plt.plot(range(100, 10000,100), t_array, c='r')
plt.plot(range(100, 10000,100), t_array1, c='b')
plt.figlegend(labels=('error', 'error1'))
plt.show()
```

По графику видим, что обрабатывать объемные данные выгоднее работая с векторами и матрицами.
